# WeRateDogs Twitter Data: Visualization Report

```python
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
import tweepy
import time
import os
sns.set_context()

plt.style.use('seaborn')
plt.figure(figsize=(12,8))

%matplotlib inline
```

```python
#import master data from data wrangling process

tweets_data_master = pd.read_csv('tweets_data_master.csv')
```

```python
ax1=sns.pairplot(tweets_data_master[['Growth_phase',"Retweets",'Day','Favorite_count','Rating']],hue='Growth_phase')
```

```python
ax2=sns.heatmap(tweets_data_master[['Growth_phase',"Retweets",'Day','Favorite_count','Rating']].corr(),annot=True)
```

#### The visual indicate there is a strong relationship between favorite count and retweets( mean dogs with high number of retweets are most likely to have high favorite counts). there is apoor relationship between rating and favorite count and also with retweets. this means high rating might not actually translate to high retweets or favorite counts.

```python
sns.catplot(x="Growth_phase",kind='count',aspect=1.5,data=tweets_data_master[tweets_data_master["Growth_phase"]!="No Name"])
```

#### The dogs in pupper phase of growth have the highest count in the tweet data.

```python
sns.catplot(x='Growth_phase',y='Favorite_count',kind='bar',aspect=1.2,ci=None,data=tweets_data_master)
```

#### Dogs in the puppo and Doggo growth phase account for the heighest favorite counts

```python
sns.catplot(x='Growth_phase',y='Retweets',kind="bar", aspect=1.2,ci=None,data=tweets_data_master)
```

#### Dogs in the puppo and Doggo growth phase account for the heighest retweets

```python
sns.catplot(x='Day', kind='count',aspect=1.5,data=tweets_data_master)
```

#### The highest number of tweets occurred on monday while the least occurred on Saturday. Tuesday, Wednesday,Thursday and Friday have almost equal tweet count.

```python
sns.catplot(x='Day',y='Retweets',ci=None,kind='bar',aspect=1.2,data=tweets_data_master)
```

#### Although most tweets were posted on Monday, highest number of retweets occurred on Wednesday.

```python
sns.catplot(x='Day',y='Favorite_count',kind='bar', aspect=1.2, ci=None,data=tweets_data_master)
```

#### Although most tweets were posted on Monday, highest number of favorite count occurred on Wednesday. The information from retweets and favorite count indicate that highest activity occurred on Wednesday.

```python
sns.catplot('Source', kind='count',data=tweets_data_master,aspect=1.5)
```

#### Majority of the tweets (over 95%) were posted from iPhone.

```python
sns.barplot(x='Source',y="Retweets",ci=None,data=tweets_data_master)
```

#### Majority of the Retweets were also posted from iPhone.

```python
sns.barplot(x='Growth_phase',y='Rating',ci=None,data=tweets_data_master[tweets_data_master["Growth_phase"]!="No Name"])
```

#### Doggo, Puppo and Floofer have also equal average rating

### Conclusion

#### The visualization indicate that having a larger number of tweets does not translate to high amount of retweets or favorite count. Both tweets and retweets originated from Twitter for iPhone, which must have been because ot the ease of tweeting and retweeting from the phone. There is also a very poor correlation between rating and favorite count or retweets. High ratings for a particular dog does not translate to high retweets or favorite count. This means there might be possiblly be more factors aside rating which are considered by individuals who visit the WeRateDogs twitter account.
